```python
df=pd.read_csv('ecommerce_sample_dataset.csv')
df
```

```python
df['crawl_timestamp'] = pd.to_datetime(df['crawl_timestamp'])
```

```python
df['crawl_year'] = df['crawl_timestamp'].apply(lambda x: x.year)
df['crawl_month'] = df['crawl_timestamp'].apply(lambda x: x.month)
```

```python
df['MainCategory'] = df['product_category_tree'].apply(lambda x: x.split('>>')[0][2:])
```

```python
def secondary(x):
    try:
        return x.split('>>')[1][1:]
    except IndexError:
        return 'None '
    
def tertiary(x):
    try:
        return x.split('>>')[2][1:]
    except IndexError:
        return 'None '
    
def quaternary(x):
    try:
        return x.split('>>')[3][1:]
    except IndexError:
        return 'None '
```

```python
df['SecondaryCategory'] = df['product_category_tree'].apply(secondary)
df['TertiaryCategory'] = df['product_category_tree'].apply(tertiary)
df['QuaternaryCategory'] = df['product_category_tree'].apply(quaternary)
df['MainCategory']=df['MainCategory'].str.strip('"]')
df['SecondaryCategory']=df['SecondaryCategory'].str.strip('"]')
df['TertiaryCategory']=df['TertiaryCategory'].str.strip('"]')
df['QuaternaryCategory']=df['QuaternaryCategory'].str.strip('"]')
df
```
